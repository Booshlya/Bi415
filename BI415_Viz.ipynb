```python
import pandas as pd
import plotly.express as px
from datetime import datetime
from collections import defaultdict
from google.colab import files

file_name = 'output_ex1.txt'
data = pd.read_csv(file_name, sep='\t', header=None, names=['Year-Month', 'Number of Ratings'])

data['Year-Month'] = pd.to_datetime(data['Year-Month'], format='%Y-%m')


fig = px.bar(data,
             x='Year-Month',
             y='Number of Ratings',
             title="Number of Ratings per Month",
             labels={'Year-Month': 'Month-Year', 'Number of Ratings': 'Ratings'},
             color='Number of Ratings',
             color_continuous_scale='Viridis')


fig.add_hline(y=12000,
              line_dash="dash",
              line_color="red",
              annotation_text="12000 Threshold",
              annotation_position="top right")


fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
#this file to load u.data and check

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

month_counts = defaultdict(int)

with open(file_name, 'r') as file:
    for line in file:
        fields = line.strip().split('\t')
        if len(fields) == 4:
            try:
                timestamp = int(fields[3])
                year_month = datetime.fromtimestamp(timestamp).strftime('%Y-%m')
                month_counts[year_month] += 1
            except ValueError:
                continue


for year_month in sorted(month_counts):
    print(f"{year_month}\t{month_counts[year_month]}")
```

```python
file_name = 'weekdays.txt'
data = pd.read_csv(file_name, sep='\t', header=None, names=["key", "count"])

data.head()


data['key'] = data['key'].apply(lambda x: x.strip())

data_sorted = data.groupby('key')['count'].sum().reset_index()

fig = px.bar(data_sorted, x='key', y='count', title="Distribution of Ratings by Day of the Week")
fig.update_layout(xaxis_title='Day of the Week', yaxis_title='Number of Ratings', xaxis_tickangle=-45)

fig.show()
```

Code to display 4a

```python
file_name_a = 'part-r-00000 (3).txt'


df_a = pd.read_csv(file_name_a, sep=',', header=None, names=['Occupation', 'Number of Ratings', 'Average Rating'])

df_a_sorted_ratings = df_a.sort_values(by='Number of Ratings')

fig_ratings_a = px.bar(
    df_a_sorted_ratings,
    x='Occupation',
    y='Number of Ratings',
    title='Number of Ratings by Occupation',
    labels={'Occupation': 'Occupation', 'Number of Ratings': 'Number of Ratings'},
    color='Occupation'
)
fig_ratings_a.update_layout(
    xaxis_title='Occupation',
    yaxis_title='Number of Ratings'
)
fig_ratings_a.show()

df_a_sorted_avg = df_a.sort_values(by='Average Rating')

fig_avg_rating_a = px.bar(
    df_a_sorted_avg,
    x='Occupation',
    y='Average Rating',
    title='Average Rating by Occupation',
    labels={'Occupation': 'Occupation', 'Average Rating': 'Average Rating'},
    color='Occupation'
)
fig_avg_rating_a.update_layout(
    xaxis_title='Occupation',
    yaxis_title='Average Rating'
)
fig_avg_rating_a.show()
```

Code to display the result of 4b

```python
file_name = 'part-r-00000 (4).txt'


df = pd.read_csv(file_name, sep=',', header=None, names=['Occupation', 'Gender', 'Number of Ratings', 'Average Rating'])

df_sorted_ratings = df.sort_values(by='Number of Ratings')

fig_ratings = px.bar(
    df_sorted_ratings,
    x='Occupation',
    y='Number of Ratings',
    color='Gender',
    barmode='group',
    title='Number of Ratings by Occupation and Gender',
    labels={'Number of Ratings': 'Number of Ratings', 'Occupation': 'Occupation'}
)
fig_ratings.update_layout(
    xaxis_title='Occupation',
    yaxis_title='Number of Ratings'
)
fig_ratings.show()
df_sorted_avg = df.sort_values(by='Average Rating')
fig_avg_rating = px.bar(
    df_sorted_avg,
    x='Occupation',
    y='Average Rating',
    color='Gender',
    barmode='group',
    title='Average Rating by Occupation and Gender',
    labels={'Average Rating': 'Average Rating', 'Occupation': 'Occupation'}
)
fig_avg_rating.update_layout(
    xaxis_title='Occupation',
    yaxis_title='Average Rating'
)
fig_avg_rating.show()
```
